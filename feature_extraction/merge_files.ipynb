```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 30)
```

```python
# 盲测集和重测盲测集
val_before = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_before.csv', index_col='id')
val_after = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_after.csv', index_col='id')
val_before_retest = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_before_retest.csv', index_col='id')
val_after_retest = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_after_retest.csv', index_col='id')

val_before_refine = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_before_refine.csv', index_col='id')
val_after_refine = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_after_refine.csv', index_col='id')
val_before_retest_refine = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_before_retest_refine.csv', index_col='id')
val_after_retest_refine = pd.read_csv('/data/gas_blind_8feat/values/RGB_val_after_retest_refine.csv', index_col='id')

val_before.index = map(lambda x: int(x[1:]), val_before.index)
val_before.sort_index(inplace=True)
val_after.index = map(lambda x: int(x[1:]), val_after.index)
val_after.sort_index(inplace=True)
val_before_refine.index = map(lambda x: int(x[1:]), val_before_refine.index)
val_before_refine.sort_index(inplace=True)
val_after_refine.index = map(lambda x: int(x[1:]), val_after_refine.index)
val_after_refine.sort_index(inplace=True)

val_after_retest.sort_index(inplace=True)
val_before_retest.sort_index(inplace=True)
val_before_retest_refine.sort_index(inplace=True)
val_after_retest_refine.sort_index(inplace=True)
```

```python
before_new = pd.concat([val_before.iloc[:20], val_before_retest])
before_new.index = range(1, len(before_new) + 1)
before_new.index.name = 'id'
after_new = pd.concat([val_after.iloc[:20], val_after_retest])
after_new.index = range(1, len(after_new) + 1)
after_new.index.name = 'id'
before_refine_new = pd.concat([val_before_refine.iloc[:20], val_before_retest_refine])
before_refine_new.index = range(1, len(before_refine_new) + 1)
before_refine_new.index.name = 'id'
after_refine_new = pd.concat([val_after_refine.iloc[:20], val_after_retest_refine])
after_refine_new.index = range(1, len(after_refine_new) + 1)
after_refine_new.index.name = 'id'
```

```python
before_new.to_csv('/data/gas_all/values/before_blind.csv')
after_new.to_csv('/data/gas_all/values/after_blind.csv')
before_refine_new.to_csv('/data/gas_all/values/before_refine_blind.csv')
after_refine_new.to_csv('/data/gas_all/values/after_refine_blind.csv')
```

```python
# 测试集
rgb_before = pd.read_csv('/data/gas_8feat/values/RGB_val_before.csv', index_col='id')
rgb_after = pd.read_csv('/data/gas_8feat/values/RGB_val_after.csv', index_col='id')
rgb_before_refine = pd.read_csv('/data/gas_8feat/values/RGB_val_before_refine.csv', index_col='id')
rgb_after_refine = pd.read_csv('/data/gas_8feat/values/RGB_val_after_refine.csv', index_col='id')

before_all = pd.concat([rgb_before, before_new], axis=0)
before_all.index = range(1, len(before_all) + 1)
before_all.index.name = 'id'
after_all = pd.concat([rgb_after, after_new], axis=0)
after_all.index = range(1, len(after_all) + 1)
after_all.index.name = 'id'
before_refine_all = pd.concat([rgb_before_refine, before_refine_new], axis=0)
before_refine_all.index = range(1, len(before_refine_all) + 1)
before_refine_all.index.name = 'id'
after_refine_all = pd.concat([rgb_after_refine, after_refine_new], axis=0)
after_refine_all.index = range(1, len(after_refine_all) + 1)
after_refine_all.index.name = 'id'
```

```python
before_all.to_csv('/data/gas_all/values/before_all.csv')
after_all.to_csv('/data/gas_all/values/after_all.csv')
before_refine_all.to_csv('/data/gas_all/values/before_refine_all.csv')
after_refine_all.to_csv('/data/gas_all/values/after_refine_all.csv')
```
